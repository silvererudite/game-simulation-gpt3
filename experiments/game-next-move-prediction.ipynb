```python
import os
from tqdm.auto import tqdm
from transformers import GPT2TokenizerFast
import openai


os.environ['OPENAI_API_KEY'] = ""
```

```python
openai.api_key = os.environ.get('OPENAI_API_KEY')
COMPLETIONS_MODEL = "text-davinci-003"
```

```python
COMPLETIONS_API_PARAMS = {
    # We use temperature of 0.0 because it gives the most predictable, factual answer.
    "temperature": 0.0,
    "max_tokens": 500,
    "n": 1,
    "stop": None,
    "model": COMPLETIONS_MODEL,
}
```

```python
prompt1 = """ Please read the game rules from this link and come up with a system to get the set of optimal next moves for a player of this game
    https://gamerules.com/rules/7-wonders-duel/
"""
```

```python
def generate_answer(prompt):
    #Pass the generated prompt and pass it to gpt-3 to get answers.
    response = openai.Completion.create(
                prompt=prompt,
                **COMPLETIONS_API_PARAMS
            )

    return response["choices"][0].text
```

```python
system_rules = generate_answer(prompt1)
```

```python
system_rules.split("\n")
```

```python
prompt2 = f"""For this game here https://gamerules.com/rules/7-wonders-duel/ there are these rules that we think of as a system {system_rules}, please come up with a set of different test cases

to test the system in the form of a list also explain the cases that you come up with.
"""
```

```python
scenarios = generate_answer(prompt2)
```

```python
scenarios.split("\n")
```

```python
prompt3 = f"""For this game here https://gamerules.com/rules/7-wonders-duel/ there are these rules that we think of as a system {system_rules}, please come
up with a game session between two players in the form of 
Player A: [Player A's move]
Player B: [Player B's move]

Based on the rules determine if player B wins or loses in 3 moves and explain your reasoning. Determine if the moves fit any scenarios from here {scenarios} state that test case
"""
```

```python
game_session = generate_answer(prompt3)
```
